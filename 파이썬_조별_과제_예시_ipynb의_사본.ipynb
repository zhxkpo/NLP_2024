```python
class Product:  #상품을 나타내는 클래스.

    def __init__(self, name, price, description):
        self.name = name
        self.price = price
        self.description = description

    def __str__(self):
        return f"{self.name} ({self.price}원): {self.description}"


class Cart:  #장바구니를 나타내는 클래스.

    def __init__(self):
        self.items = []

    def add_item(self, product, quantity):
        self.items.append((product, quantity))

    def remove_item(self, product):
        self.items.remove(product)

    def __len__(self):
        return len(self.items)

    def __str__(self):
        return "\n".join([f"{product} x {quantity}" for product, quantity in self.items])


class Order:  #주문을 나타내는 클래스.

    def __init__(self, cart, customer):
        self.cart = cart
        self.customer = customer

    def __str__(self):
        return f"{self.customer.name}님의 주문: {self.cart}"


class Customer:  #온라인 쇼핑몰 계정을 나타내는 클래스.

    def __init__(self, user_id, password, name, address, email):
        self.user_id = user_id
        self.password = password
        self.name = name
        self.address = address
        self.email = email
        self.order_history = []  # 주문 내역 저장

    def add_order(self, order):  #주문 내역에 주문을 추가.
        self.order_history.append(order)

    def view_order_history(self):  #주문 내역을 출력.
        print(f"{self.name}님의 주문 내역:")
        for order in self.order_history:
            print(str(order))

    def __str__(self):
        return f"아이디: {self.user_id}, 이름: {self.name}, 주소: {self.address}, 이메일: {self.email}"


class OnlineShoppingMall:  #온라인 쇼핑몰을 나타내는 클래스.

    def __init__(self):
        self.products = []
        self.customers = []

    def add_product(self, product):
        self.products.append(product)

    def add_customer(self, customer):
        self.customers.append(customer)

    def __str__(self):
        return f"상품: {self.products}\n고객: {self.customers}"

customer1 = Customer("user123", "password456", "김철수", "서울시 강남구", "user123@example.com")



product1 = Product("티셔츠", 20000, "면 100% 티셔츠")
product2 = Product("바지", 30000, "데님 바지")
product3 = Product("신발", 50000, "운동화")
product4 = Product("모자", 10000, "야구 모자")
product5 = Product("가방", 40000, "백팩")


# 주문 생성
cart = Cart()
cart.add_item(product1, 2)
cart.add_item(product3, 1)
order = Order(cart, customer1)

# 고객의 주문 내역에 주문 추가
customer1.add_order(order)

# 고객 정보 출력
print(customer1)

# 주문 내역 출력
customer1.view_order_history()
```

```python
class Product:
    """상품을 나타내는 클래스입니다."""

    def __init__(self, name, price, description):
        self.name = name
        self.price = price
        self.description = description

    def __str__(self):
        return f"{self.name} ({self.price}원): {self.description}"


class Cart:
    """장바구니를 나타내는 클래스입니다."""

    def __init__(self):
        self.items = []

    def add_item(self, product, quantity):
        self.items.append((product, quantity))

    def remove_item(self, product):
        self.items.remove(product)

    def __len__(self):
        return len(self.items)

    def __str__(self):
        return "\n".join([f"{product} x {quantity}" for product, quantity in self.items])


class Order:
    """주문을 나타내는 클래스입니다."""

    def __init__(self, cart, customer):
        self.cart = cart
        self.customer = customer

    def __str__(self):
        order_details = []
        for product, quantity in self.cart.items:
            order_details.append(f"{product.name} ({product.price}원): {product.description} x {quantity}")
        return f"{self.customer.name}님의 주문:\n" + "\n".join(order_details)


class Customer:
    """온라인 쇼핑몰 계정을 나타내는 클래스입니다."""

    def __init__(self, user_id, password, name, address, email, balance=0):
        self.user_id = user_id
        self.password = password
        self.name = name
        self.address = address
        self.email = email
        self.order_history = []  # 주문 내역 저장
        self.balance = balance  # 계정 잔액 추가

    def add_balance(self, amount):
        """계정에 잔액을 추가합니다."""
        self.balance += amount

    def place_order(self, order):
        """주문을 합니다. 잔액이 부족하면 주문에 실패합니다."""
        total_price = sum(product.price * quantity for product, quantity in order.cart.items)
        if self.balance >= total_price:
            self.balance -= total_price
            self.add_order(order)
            print("주문이 완료되었습니다.")
        else:
            print("잔액이 부족하여 주문에 실패했습니다.")

    def add_order(self, order):
        """주문 내역에 주문을 추가합니다."""
        self.order_history.append(order)

    def view_order_history(self):  # 주문 내역을 출력합니다.
        print(f"{self.name}님의 주문 내역:")
        for order in self.order_history:
            # order는 Order 객체이므로, Order 클래스의 __str__ 메소드를 호출하여 출력합니다.
            print(str(order))  # 또는 print(order)

    def __str__(self):
        return f"아이디: {self.user_id}, 이름: {self.name}, 주소: {self.address}, 이메일: {self.email}"


class OnlineShoppingMall:
    """온라인 쇼핑몰을 나타내는 클래스입니다."""

    def __init__(self):
        self.products = []
        self.customers = []

    def add_product(self, product):
        self.products.append(product)

    def add_customer(self, customer):
        self.customers.append(customer)

    def __str__(self):
        return f"상품: {self.products}\n고객: {self.customers}"


# --- 상품 객체 생성 ---
product1 = Product("티셔츠", 20000, "면 100% 티셔츠")
product2 = Product("바지", 30000, "데님 바지")
product3 = Product("신발", 50000, "운동화")
product4 = Product("모자", 10000, "야구 모자")
product5 = Product("가방", 40000, "백팩")
```

```python
# 온라인 쇼핑몰 객체 생성
mall = OnlineShoppingMall()

# 상품 객체 생성 및 쇼핑몰에 추가
product1 = Product("티셔츠", 20000, "면 100% 티셔츠")
product2 = Product("바지", 30000, "데님 바지")
product3 = Product("신발", 50000, "운동화")
mall.add_product(product1)
mall.add_product(product2)
mall.add_product(product3)

# 고객 객체 생성 및 쇼핑몰에 추가
customer1 = Customer("user123", "password456", "김철수", "서울시 강남구", "user123@example.com", balance=100000)
customer2 = Customer("user456", "pass789", "홍길동", "경기도 수원시", "user456@example.com", balance=50000)
mall.add_customer(customer1)

# --- 고객1의 쇼핑 ---
# 장바구니 객체 생성 및 상품 추가
cart1 = Cart()
cart1.add_item(product1, 2)  # 티셔츠 2개 추가
cart1.add_item(product3, 1)  # 신발 1개 추가

# 주문 객체 생성 및 주문하기
order1 = Order(cart1, customer1)
customer1.place_order(order1)  # 주문 완료

# --- 고객2의 쇼핑 ---
# 장바구니 객체 생성 및 상품 추가
cart2 = Cart()
cart2.add_item(product2, 1)  # 바지 1개 추가
cart2.add_item(product3, 2)  # 신발 2개 추가

# 주문 객체 생성 및 주문하기
order2 = Order(cart2, customer2)
customer2.place_order(order2)  # 잔액 부족으로 주문 실패

# 고객1의 주문 내역 확인
customer1.view_order_history()

# 고객2의 잔액 추가 및 재주문
customer2.add_balance(100000)
customer2.place_order(order2)  # 주문 완료

# 고객2의 주문 내역 확인
customer2.view_order_history()
```

```python
import tutle as t # 화재 경보

tempc=50
alert=['정상','주의','위험']
color=['white','blue','red']

def check():
    if tempc<80:
        status=0
    elif tempc<120:
        status=1
    else:
        status=2

    t.clear()
    t.home()
    t.down()
    t.fillcolor(color[status])
    t.begin_fill()
    t.circle(20)
    t.end_fill()
    t.up()

    t.goto(0,50)
    tmp=alert[status]+str(tempc)
    t.write(tmp,align='center')

def k_up90:
    global tempc
    if tempc<80:
        tempc+=5
    else:
        tempc+=10
    check()

def key_down():
    global tempc
    if tempc<=80:
        tempc-=5
    else:
        tempc-=10
    check()

t.setup(400,400)
t.hideturtle()
t.speed(0)
t.up()
check()

s=t.Screen()
s.onkey(k_up,'Up')
s.onkey(k_down,'Down')
s.onkey(s.bye,'q')
s.listen()
```
